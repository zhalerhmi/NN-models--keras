```python
import pandas as pd
import numpy as np
```

```python
#load data
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
#split predictors and target
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
#normalize data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
    return model
```

```python
#split data train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
```

```python
# build the model
model = regression_model()
```

```python
# fit the model using data train
model.fit(X_train, y_train, epochs=100, verbose=2)

```

```python
# evaluate the model using data test
scores = model.evaluate(X_test, y_test, verbose=0)
print("MSE: {}".format(scores[0]))
```

```python
# evaluate the model using data test with sklearn 
from sklearn.metrics import mean_squared_error
y_pred= model.predict(X_test)

print("MSE: {}".format(mean_squared_error(y_test, y_pred)))
```

```python
#Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.
MSE_scores=[]

for i in range(1,50):
        X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3) #split data
        model = regression_model()
        model.fit(X_train, y_train, epochs=100, verbose=0) #train model on training data using 50 epoc
        scores = model.evaluate(X_test, y_test, verbose=1)
        MSE_scores.append(scores[0])



```

```python
mean_MSE=np.mean(MSE_scores)
STD_MSE=np.std(MSE_scores)

#Report the mean and the standard deviation of the mean squared errors.
print('mean_MSE: {} \n STD_MSE: {}'.format(mean_MSE, STD_MSE))
        
```
